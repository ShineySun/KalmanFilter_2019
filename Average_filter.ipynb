### 평균 필터는 하나씩 차례로 입력되는 데이터의 평균을 계산할 때 유용.
### 평균 필터는 센서 초기화에 사용.
### 전원을 키고 처음에 일정 시간 동안 센서의 출력값을 모아서 평균을 내고, 이 값을 영점으로 잡는 초기화 작업이 필요.

```
### 예제 코드
### 10개의 1 ~ 10 까지의 랜덤 정수를 순서대로 평균 필터링함.

import numpy as np

# pre_avg value
pre_avg = 0
# first_run flag
first_run = True
# num of data 
k = 0

def AvgFilter(x):
  global pre_avg
  global k
  global first_run
  
  if(first_run):
    pre_avg = x
    first_run = False
    k = k+1

    return pre_avg
  k = k+1
  alpha = (k-1)/k
  result = alpha*pre_avg + (1-alpha)*x

  pre_avg = result
  

  return result

# main
random_list = np.random.randint(0,10,10)
print("Random Data : ",random_list)

for i,x in enumerate(random_list):
  avg = AvgFilter(x)
  print('index : {}, avg : {}'.format(i,avg))


  





```
